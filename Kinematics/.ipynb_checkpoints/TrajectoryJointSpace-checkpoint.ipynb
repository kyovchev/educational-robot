```python
import roboticstoolbox as rtb
from spatialmath import SE3
import numpy as np
import os
import swift
import time
```

## Load Robot from URDF

```python
robot = rtb.ERobot.URDF(os.path.abspath("robot.urdf"))

print(robot)
for i, link in enumerate(robot.links):
    if link.isjoint:
        ql = link.qlim
        print(f"Joint {i} ({link.name}): [{np.degrees(ql[0]):.1f}°, {np.degrees(ql[1]):.1f}°]")
```

## Trajectory in Joint Space

```python
start = SE3(0.12, 0.12, 0.09) * SE3.RPY(np.pi, 0, 0)
end   = SE3(0.12, 0.12, 0.01) * SE3.RPY(np.pi, 0, 0)

sol = robot.ikine_LM(start, q0=np.zeros(robot.n), tol=0.001, seed=100)
q_start = sol.q
sol = robot.ikine_LM(end, q0=q_start, tol=0.001, seed=100)
q_end = sol.q

traj = rtb.jtraj(q_start, q_end, 50)

env = swift.Swift()
env.launch(browser='notebook')
env.add(robot)

for _ in range(10):
    for q in traj.q:
        robot.q = q
        env.step(0.05)
        time.sleep(0.05)
```
